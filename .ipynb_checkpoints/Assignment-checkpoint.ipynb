```python
image = cv2.imread('./spongebob_wide-56a00f0a5f9b58eba4aeb6f2.jpg')
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.title('Spongebob')
plt.imshow(image, interpolation='none')
```

## Making array of pixelcolors

```python
image = img.imread('./spongebob_wide-56a00f0a5f9b58eba4aeb6f2.jpg')
print("Shape of image: ", image.shape)
size = image.shape[0]*image.shape[1]
print("Total amount of pixels in image: ", size)
```

```python
r = []
g = []
b = []
for line in image:
    for pixel in line:
        temp_r, temp_g, temp_b = pixel
        r.append(temp_r)
        g.append(temp_g)
        b.append(temp_b)

df = pd.DataFrame({'red': r,
'blue': b,
'green': g})
```

```python
npa = df.values
```

```python
npa
```

```python
uniquearray = np.unique(npa, axis=0)
```

## Invert colors of image

```python
invert_raw = PIL.Image.open('./spongebob_wide-56a00f0a5f9b58eba4aeb6f2.jpg')
invert_done = PIL.ImageOps.invert(invert_raw)
invert_done.save('spongebob_inverted.jpg')
```
